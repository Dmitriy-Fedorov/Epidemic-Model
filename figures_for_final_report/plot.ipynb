```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csv = 'monte_random_r=2_15.csv'
pdf_title = '{}.png'.format(csv.split('.')[0])
asd = pd.read_csv(csv)
asd1 = asd[['I1_a', 'I1_s','I2_a','I2_s']]/900
f = asd1[['I1_a', 'I1_s','I2_a','I2_s']].plot(grid=True, figsize=[15,9])
plt.title('plot_title')
plt.xlabel('time(t)')
plt.ylabel('Density')
plt.xlim([0, 100])
# plt.ylim([-0.05,1.05])
f.get_figure().savefig(pdf_title, bbox_inches='tight')
```
