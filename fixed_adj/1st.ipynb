```python
def monte(n, r, paramet, nstep=100, I1_a=10, I2_a=10, grid_size=[30,30], net='Geometric Random'): # 'Geometric Random', 'Uniform Grid'
    for item in locals().items():
        print(item)
    listt = []
    for i in range(n):
        a = EpidemicGraph(N, r, paramet, I1_a=I1_a, I2_a=I1_a, grid_size=grid_size, net=net)
        a.run(nstep)
        listt.append(a.hist2pandas())
        number_subgraphs = len(list(nx.connected_component_subgraphs(a.G)))
        number_isolated_nodes = len(list(nx.isolates(a.G)))
        print(f'({i+1} {number_isolated_nodes}/{number_subgraphs} {calc_R0(a.G, paramet)})', end='\n')
    print()
    df = listt[0]
    del listt[0]
    for v in listt:
        df += v
    df /= n
    return df

def monte_load(n, a, nstep): # 'Geometric Random', 'Uniform Grid'
    for item in locals().items():
        print(item)
    listt = []
    for i in range(n):
        a.run(nstep)
        listt.append(a.hist2pandas())
        number_subgraphs = len(list(nx.connected_component_subgraphs(a.G)))
        number_isolated_nodes = len(list(nx.isolates(a.G)))
        print(f'({i+1} {number_isolated_nodes}/{number_subgraphs} {calc_R0(a.G, paramet)})', end='\n')
    print()
    df = listt[0]
    del listt[0]
    for v in listt:
        df += v
    df /= n
    return df, listt

N = 900
r=1.5
paramet = {
        'alpha': [0.25, 0.35], # infect rate 
        'mu': [0.04, 0.04], # sleep s 
        'gamma': [0.04, 0.35], # rec rate 
        'lambda': [0.06, 0.07], # sleep I1 
        'kappa': [0.05, 0.06] # sleep I2    % [I2_s -> I2_a, I2_a -> I2_s]
    }


setup = {'n':5, 
         'r':1.5, 
         'paramet': paramet,
         'nstep':300,
         'I1_a':50, 
         'I2_a':50, 
         'grid_size':[30,30]}
a = EpidemicGraph.load(N,r,paramet,I1_a=50, I2_a=50, grid_size=[30,30]) # net='Uniform Grid'



# rnd_monte = monte(net='Geometric Random', **setup)
rnd_monte, listt = monte_load(15, a, nstep=300)
rnd_monte.to_csv(f"monte_random_r={setup['r']}_{setup['n']}.csv", index=False)
asd1 = rnd_monte[['I1_a','I1_s','I2_a','I2_s']]
asd1.plot()


```
